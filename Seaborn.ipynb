```python
#Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
```

```python
#numerical data plotting
#relplot has options of plotting both categorical and numerical data
#relplot>>how the relationship between two variables
#relplot>> relationship plot
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = df)
```

```python
sns.relplot(x = 'total_bill', y='tip', data = df, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y='tip', data = df, style = 'size')
```

```python
sns.relplot(x = 'total_bill', y='tip', data = df, style = 'smoker', hue = 'time')
```

```python
#relplot for making a line plot>> generating a data
#line plot>> to understand the trend in data, a time column/sequence is generally is there
```

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
df = pd.DataFrame({'time': time, 'value': value})
df
```

```python
sns.relplot(x = 'time', y ='value', data = df, kind = 'line')
```

```python
sns.relplot(x = 'time', y ='value', data = df, kind = 'line', sort = True)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker')
```

```python
sns.relplot(data = tips, x = 'total_bill', y = 'tip', col = 'size', hue = 'smoker')
```

```python
#for categorical column>> cat plot
#plot between day and total bill, day is a categorical column
sns.catplot(x = 'day', y ='total_bill', data = tips)
```

```python
sns.catplot(x = 'day', y ='total_bill', data = tips, kind = 'swarm') # to prevent the overlapping of points
```

```python
#boxplot : to see outliers
sns.catplot(x = 'day', y ='total_bill', data = tips, kind = 'box')
```

```python
sns.catplot(x = 'day', y ='total_bill', data = tips, kind = 'violin')
```

```python
sns.catplot(x = 'day', y ='total_bill', data = tips, kind = 'bar')
```

```python
sns.barplot(x = 'day', y ='total_bill', data = tips)
```

```python
#draw a plot of two variables with distribution>>bivariate plots
# showing the relationship between 2 variables as well as their individual distribution
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')# regressio line to represent relationship
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde') #kernel density estimator, contour>>meterological
```

```python
sns.pairplot(tips)  #make plots scatter plot with respect to all numerical columns
```

```python
sns.pairplot(tips, hue = 'smoker')
```

```python
sns.pairplot(tips, hue = 'smoker', palette = 'coolwarm')
# https://seaborn.pydata.org/generated/seaborn.color_palette.html
# https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'box', palette = 'rainbow')
#
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, palette = 'rainbow')
```
