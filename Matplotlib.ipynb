```python
x = np.random.randn(50)
y = np.random.randn(50)
```

```python
#scatter plot >> A scatter plot is a type of data visualization that displays values for two variables as points on a Cartesian plane. Each point represents the value of one variable corresponding to the value of the other variable
plt.scatter(x,y, color='r')
```

```python
#line plot
#whenever you want to depict the relation between two variables in the form of trend
#mostly in x axis the date - time column is there
x = [1, 2, 3, 4, 5] #assume x is no of day
y = [5, 2, 7, 8, 1] #y is price of stock
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, color='g')
plt.title("Line Chart")
plt.xlabel("No of Days")
plt.ylabel("Proce of the stock")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, color='g', marker = 'o', linestyle = '--')
plt.title("Line Chart")
plt.xlabel("No of Days")
plt.ylabel("Proce of the stock")
plt.grid()
plt.show()
```

```python
plt.plot(x,y, color='g', marker = 'o', linestyle = '--', linewidth = 4, markersize = 10)
plt.title("Line Chart")
plt.xlabel("No of Days")
plt.ylabel("Proce of the stock")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [5, 1, 7, 8, 2]
y2 = [7, 8, 3, 5, 1]
plt.plot(x,y1,color='r', label = 'line 1', marker = 'o', linestyle = '--')
plt.plot(x, y2, color = 'b', label = 'line 2', marker = 'o', linestyle = '--')
plt.grid()
plt.legend()
plt.title("Stakced Line Chart")
plt.xlabel("No of days")
plt.ylabel("Price of stock")
plt.show()
```

```python
plt.figure() #initialising a page to make plot
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.figure() #initialising a page to make plot
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.show()
```

```python
plt.figure() #initialising a page to make plot
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
plt.subplot(2,2,4)
plt.plot(x, np.cos(x+1))
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'magenta')
plt.xlim(1,15)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
plt.plot(x, np.sin(x), color = 'magenta')
plt.xlim(1,15)
plt.ylim(-1.5, 1.5)
plt.axis('equal')
plt.grid()
plt.title("basic syntax")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
#histogram>> continuous numerical data
x = np.random.randn(1000)
```

```python
plt.hist(x)
plt.show() #be default it shows 10 bins we can explicitly specify the number of bins here
```

```python
plt.hist(x, bins = 20, color = 'lightgreen')
plt.show()
```

```python
data1 = np.random.randn(50)
data2 = np.random.randn(50)
plt.hist([data1, data2], bins = 10, color = ['r','b'], label = ['Data1', 'Data2'])
plt.xlabel("Values")
plt.ylabel('frequency')
plt.legend()
plt.title("stacked histogram")
```

```python
#scatterplot >> to understand the relationship between two features
plt.scatter(data1, data2, marker = 'X')
plt.show()
```

```python
#bar plot >> To see the districution of categorical variables
x = ["a", "b", "c","d"] #name of students
y = [4, 6, 1, 10] #no of days present in the class

plt.bar(x,y, color = 'magenta')
plt.show()
```

```python
plt.barh(x,y, color = 'magenta')
plt.show()
```

```python
#pie char >> To understand the percent distribution of data
langs = ["C", "Python", "Java"]
students = [20, 100, 40]
plt.pie(students, labels = langs, autopct = "%1.1f%%")
plt.show()
```

```python
explode = [0,0.1, 0.1]
plt.pie(students, labels = langs, autopct = "%1.1f%%",explode = explode,shadow = True)
plt.show()
```
